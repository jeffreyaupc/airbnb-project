# Load data

```python
import pandas as pd

df = pd.read_csv('listings.csv', low_memory=False)

print(df.shape)
```

# Check columns with 100% missing values

```python
total_rows = df.shape[0]
total_missing = df.isnull().sum()

total_missing[total_missing == total_rows].index
```

# Drop these columns

```python
new_df = df.drop(columns=total_missing[total_missing == total_rows].index, axis=1)

print(new_df.shape)
```
