```python
import pandas as pd
import ast
import plotly.express as px

df = pd.read_csv('listings_clean.csv', low_memory=False)

df.shape

key_df = df[['host_name', 'neighbourhood_cleansed', 'bedrooms', 'review_scores_rating', 'amenities', 'price', 'availability_365']]
```

```python
key_df[['neighbourhood_cleansed','price']].groupby('neighbourhood_cleansed').mean().sort_values('price')
```

```python
px.bar(key_df[['neighbourhood_cleansed','price']].groupby('neighbourhood_cleansed').mean().sort_values('price').reset_index(), x='neighbourhood_cleansed', y='price', title='Average price per night by neighbourhood', labels={'price':'Price per night','neighbourhood_cleansed':'Neighbourhood'}).update_layout(xaxis_tickangle=-45)
```
